```python
# 7. Phường nào có biến động dân số nhanh nhất
data['volatility'] = data['Pop_2019'] -data['Pop_2009']
```

```python
data_volatility_max = data[data.volatility == max(data.volatility)]
data_volatility_max
```

Phường Vĩnh Lộc A ở Bình Chánh có biến động dân số nhanh nhất

```python
# 8. Phường nào có biến động dân số chậm nhất
data_volatility_min = data[data.volatility == min(data.volatility)]
data_volatility_min
```

```python
# 9. Phường nào có mật độ dân số cao nhất (2019)
data['population_density'] = data['Pop_2019'] / data['Shape_Area']
```

```python
data_population_density_max = data[data.population_density == max(data.population_density)]
data_population_density_max
```
