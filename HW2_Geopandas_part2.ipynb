## Create folium map

```python
m = folium.Map(location=[10.8, 106.7], zoom_start=10, tiles='CartoDB positron')
m
```

## Add polygons to map

```python
for _, r in data_dist.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
m
```

## Add centroid markers

```python
# Project to NAD83 projected crs
data_dist = data_dist.to_crs(epsg=2263)

# Access the centroid attribute of each polygon
data_dist['centroid'] = data_dist.centroid
```

```python
# Project to WGS84 geographic crs

# geometry (active) column
data_dist = data_dist.to_crs(epsg=4326)

# Centroid column
data_dist['centroid'] = data_dist['centroid'].to_crs(epsg=4326)

data_dist.head()
```

```python
for _, r in data_dist.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(m)

m
```
